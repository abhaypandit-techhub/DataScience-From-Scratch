```python
class Employee:
    company="aPAI"

    def __init__(self,owner,investment):
        self.owner=owner
        self.investment=investment
    def mycompany(self):
        print("AI revolution")
        print(f"my company is {self.company} {self.owner} and {self.investment} ")
    @staticmethod #no self parameter is needed
    def mycompany_2():
        print("fastest growing company")
    @classmethod   
    def mycompany_3(cls):
        print(f" my company is { cls.company}")#cls.name is not applicable (not in class)
obj=Employee("abhay","1000M") 
obj.company

    
```
