# Memory efficient-NumPy vs list

```python
import sys
lst_data=list(range(1000))
numpy_data=np.array(lst_data)
print("python list size:",sys.getsizeof(lst_data)*len(lst_data),"bytes")
print("NumPy array size:",numpy_data.nbytes,"bytes")
```

```python
import time
lst=list(range(1000))
arr=np.array(list(range(1000)))
start=time.time()
lst_square=[x**2 for x in lst]
end=time.time()
print("time in list :",end-start)

start=time.time()
numpy_square=arr**2
end=time.time()
print("time in NumPy :",end-start)
```
