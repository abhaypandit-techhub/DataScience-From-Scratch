# SPEED TESTING USING PYTHON LIST


```python

import time

size=10000000

l1=list(range(size))
l2=list(range(size))

start=time.time()
add=[x+y for x,y in zip(l1,l2)]
end=time.time()

print("time :",end-start)


```

 # USING NUMPY


```python
import numpy as np
import time

size=10000000

l1=np.array(list(range(size)))
l2=np.array(list(range(size)))

start=time.time()
add=l1+l2
end=time.time()
print("time :",end-start)
```
