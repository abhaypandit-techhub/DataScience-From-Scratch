```python
import pandas as pd
```

1.) CREATING DATAFRAME USING LIST

```python
data=[["Abhay",85],["Ram",69],["Pankaj",90],["shubham",82]]
```

```python
data
```

```python
pd.DataFrame(data) # without data label
```

```python
pd.DataFrame(data,columns=["Name","marks"]) # with data label
```

2.) USING DICTIONARY FRAMING DATA IN DATA FRAME

```python
data_2={"Name":["A","B","C"],
        "Age":[23,25,27]
       }
```
