# Reading excel file

```python
df=pd.read_excel("BollywoodData.xlsx")
```

```python
df=df[df["IMDb"]>6.0]
```

```python
df
```

# Writing a data into excel

```python
# update ?
#-- IMDb>6.0 file get updated
df.to_excel("data_updates.xlsx") 
```

```python
# use when no need of index
# always make sure old excel/csv file close/saved
df.to_excel("data_updates.xlsx",index=False) # here index get remove
```

| File Type     | Read Method         | Write Method      |
| ------------- | ------------------- | ----------------- |
| CSV           | `pd.read_csv()`     | `df.to_csv()`     |
| Excel (.xlsx) | `pd.read_excel()`   | `df.to_excel()`   |
| JSON          | `pd.read_json()`    | `df.to_json()`    |
| HTML Table    | `pd.read_html()`    | `df.to_html()`    |
| XML           | `pd.read_xml()`     | `df.to_xml()`     |
| SQL Database  | `pd.read_sql()`     | `df.to_sql()`     |
| Parquet       | `pd.read_parquet()` | `df.to_parquet()` |
| Pickle        | `pd.read_pickle()`  | `df.to_pickle()`  |
