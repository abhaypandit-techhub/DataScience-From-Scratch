```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
# pd.to_datetime(["2020-01-10", "2020-02-15"]) # converting dateString into actual dateTime formet
```

```python
df
```

```python
df.groupby("Department")["Salary"].mean() # it is a type of series 
```

```python
df.groupby("Gender")["Salary"].mean()
```

```python
type(df.groupby("Department")["Salary"].mean())
```

# .agg() and .aggregate() both are same

# Example :-

```python
df.groupby("Gender")["Salary"].agg(["mean","max","min"])
```

```python
df.groupby("Department")["Salary"].agg(["mean","max","min"])
```

```python
df.groupby("Department")["Salary"].aggregate(["mean","max","min"])
```

# same

```python
df.groupby("Department")["Salary"].count()
```

```python
df.groupby("Department")["Salary"].agg(
    Average_Salary="mean",Highest_Salary="max") #
```

```python
df["TeamAvg"] = df.groupby("Team")["Salary"].transform("mean")
```
