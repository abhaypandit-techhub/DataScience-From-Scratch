```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
employees
```

```python
departments
```

```python
# on is parameter telling about on which basis you merge the data
# by default here how = inner
pd.merge(employees,departments,on="DeptID") # inner join
```

```python
# This is left join , to preserve all row of data frame employees
pd.merge(employees,departments,on="DeptID",how="left") 
```

```python
# This is right join , to preserve all row of data frame departments
pd.merge(employees,departments,on="DeptID",how="right") # right join
```

```python
# Here all id is considered , it gives all row of ememployees and departments
# and where data is not clr on that place NaN 
pd.merge(employees,departments,on="DeptID",how="outer") # outer join
```

# Concatenating DataFrame

```python
df1=pd.DataFrame({"Name":["Abhay","Alok"]})
df2=pd.DataFrame({"Name":["Ram","Rohan"]})
```

```python
pd.concat([df1,df2]) # by default horizontal concatenate , here axis=0
```

```python
pd.concat([df1,df2],axis=1) # vertical concatenate
```

```python
df1=pd.DataFrame({"Name":["Abhay","Alok"],"Age":[18,19]})
df2=pd.DataFrame({"Name":["Ram","Rohan"],"Grade":["A","A+"]})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```
