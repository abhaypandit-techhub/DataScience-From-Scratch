| Column     | Meaning                                                                                    |
| ---------- | ------------------------------------------------------------------------------------------ |
| **id**     | Player ka unique ID                                                                        |
| **name**   | Player ka naam                                                                             |
| **fed**    | Federation/Country code (IND = India, BAN = Bangladesh, MAS = Malaysia)                    |
| **sex**    | Gender (M = Male, F = Female)                                                              |
| **title**  | Chess title (GM = Grandmaster, IM = International Master, FM = FIDE Master, NA = No title) |
| **wtitle** | Women's title (WGM, WIM, WFM, etc.)                                                        |
| **otitle** | Other official titles (agar koi ho)                                                        |
| **foa**    | FIDE Online Arena title (Arena Grandmaster, Arena Master, etc.)                            |
| **rating** | Chess rating (jitni zyada, utna strong player)                                             |
| **games**  | Rated games ki sankhya                                                                     |
| **k**      | K-factor (rating kitni tezi se change hogi)                                                |
| **bday**   | Birth year                                                                                 |
