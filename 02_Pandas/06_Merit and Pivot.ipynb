```python
import pandas as pd
```

# Sample DataFrame

```python
data={
    "Name":["Abhay","Kunal","Ram","Rohan","Alok"],
    "Math":[96,89,91,84,86],
    "Physic":[86,81,88,81,79],
    "English":[82,79,83,75,75],
}

df=pd.DataFrame(data)
```

```python
df
```

# -- melt() method

Syntax : df.melt(id_vars=None, value_vars=None, var_name=None, value_name="value", col_level=None)

```python
df2=df.melt(id_vars=["Name"], value_vars=["Math", "Physic", "English"], 
            var_name="Subject", value_name="Score")
```

```python
df2
```

# -- pivot()
Syntax :df.pivot(index=None, columns=None, values=None)

```python
df2.pivot(index="Name", columns="Subject", values="Score")
```
