```python
import matplotlib.pyplot as plt
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 90, 12, 24, 19, 33, 20, 15,
    94, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 96, 15, 99, 25, 40, 33, 12,
    93, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 98, 18, 38, 43, 29, 40, 93, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 93, 44, 24, 21, 24, 39, 33, 24, 95,
    30, 39, 22, 26, 26,95, 32, 32, 30, 27
]

bins = [10, 20, 30, 40, 50, 60, 70]

plt.hist(ages, bins=bins, edgecolor='black')
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.6)#alpha tell tranparency of grid
plt.show()
```

# Adding vertical line|

```python
plt.hist(ages, bins=bins, edgecolor='black')
plt.axvline(45,color='red',linestyle='-',label='Average age') # axvline = Axis Vertical Line
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```
