```python
import matplotlib.pyplot as plt
```

# scatter plot
syntax : plt.scatter(x, y)

| Parameter | Meaning                                   |
| --------- | ----------------------------------------- |
| `s`       | Size of points                            |
| `c`       | Color                                     |
| `marker`  | Shape (`'o'`, `'^'`, `'s'`, `'*'`, `'x'`) |
| `alpha`   | Transparency (0–1)                        |
| `label`   | Legend name                               |


Notes : [value_if_true if condition else value_if_false for item in iterable]
[expression for item in iterable] # In list comprehensive for comes always after expression 

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 93, 60, 65, 75, 85, 90]

sizes=[score for score in exam_scores]
colors = ['red' if score<50 else 'green' for score in exam_scores]

plt.scatter(study_hours,exam_scores,c=colors,s=sizes)
plt.xlabel("Study_hours")
plt.ylabel("Exam_scores")
plt.title("Study vs Marks")
plt.grid()
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 93, 60, 65, 75, 85, 90]

# differnt score diffrent color,when we use (cmp : colormap)
# in that case value of c is number,on changes score ,color is also changes

plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='magma')
plt.colorbar(label='score')
plt.xlabel("Study_hours")
plt.ylabel("Exam_scores")
plt.title("Study vs Marks")
plt.grid()
# https://matplotlib.org/stable/users/explain/colors/colormaps.html # this is colormap of scater plot in matplotlib
```

# Adding annotations
# synatax : plt.annotate("text", (x, y))

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 93, 60, 65, 75, 85, 90]
score=[100,200,300,400,500,600,700,800,900]
 
plt.scatter(study_hours,exam_scores,c=score,cmap='Reds')
plt.colorbar(label='score')

for i in range(len(study_hours)):
    plt.annotate(f"S {i+1}",(study_hours[i]-0.10,exam_scores[i]+0.78))
    
plt.xlabel("Study_hours")
plt.ylabel("Exam_scores")
plt.title("Study vs Marks")
plt.grid()
```

# Multiple group in single plot

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [85, 75, 65, 90]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
```

```python
plt.scatter(class_a_hours,class_a_scores,label='Class A',color='blue')
plt.scatter(class_b_hours,class_b_scores,label='Class B',color='green')
plt.grid()
plt.xlabel("Class Hours")
plt.ylabel("Class Score")
plt.title("Hours vs score")
plt.legend()
```
