# Try making a stackplot with your own weekly schedule!

```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
tution=[6,5,4,6,5,6,0]
selfstudy=[5,7,6,7,6,5,9]
playing=[2,1,3,1,1,2,2]
waching_tv=[1,1,0.3,0.5,1,1.5,2]
excercise=[1,0.30,0.50,1,0.50,1,0.30]
labels=['tution','selfstudy','playing','waching_tv','excercise']
plt.stackplot(days,tution,selfstudy,playing,waching_tv,excercise,labels=labels,alpha=0.93,edgecolor='black')
plt.xlabel("Days")
plt.ylabel("Hours")
plt.legend(loc='best')
plt.title("Weekly schedule")
plt.grid()
```
