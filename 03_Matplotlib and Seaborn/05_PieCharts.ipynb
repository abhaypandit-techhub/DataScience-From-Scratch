```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use("seaborn-v0_8")
# Data
labels1 = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.pie(runs,labels=labels1)
plt.title("Career run score of indian player")
```

```python
colors1 = ['#FC000D','#2AD4A8','#000DFC','#ffcc99'] # this is html code colors
plt.pie(runs,labels=labels1,colors=colors1)
```

```python
plt.pie(
    runs,
    labels=labels1,
    colors=colors1,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': '-'}
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels1,
    colors=colors1,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '-','fill':False}#
)
plt.show()
```

```python
explode = [0.15, 0, 0, 0]  # Only highlight Sachin's slice
# 0 represent : slice remain its original position
# 0.1 rpresent : slice comes from center position

plt.pie(runs, labels=labels1, explode=explode, colors=colors1)

wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '-'}#
plt.title("Exploded Pie Example")
plt.show()
```
