```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.lineplot(x='tip',y='total_bill',data=tips,hue='time') # hue :Color = Lunch((blue)/Dinner(orange)
plt.title("tips vs total_bill")
plt.grid()
plt.tight_layout()
```

 # bar plot

```python
sns.barplot(x='tip',y='total_bill',data=tips)
plt.title("tips vs total_bill")
plt.tight_layout()
```

# scatter plot

 # hue is used to represent an additional variable through colors, allowing us to compare different categories or numeric ranges within the same plot.

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='sex') # hue :Color = Lunch((blue)/Dinner(orange)
plt.title("tips vs total_bill")
plt.grid()
plt.tight_layout()
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,size='size')
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='smoker')
```

# box plot

```python
sns.boxplot(x='tip',y='total_bill',data=tips)
plt.title("tips vs total_bill")
plt.tight_layout()
```

```python
flights=sns.load_dataset("flights")
```

pivot use when data in long formet

```python
pv
```

# Heatmap
 -- for heatmap draw data must be needed in the form of row x column forms
 
 --  Heatmap is a graphical representation of data where individual values are represented using colors, making it easy to identify patterns, trends, and correlations in large datasets.

 -- small value : lighter color

 -- bigger value : dark color

```python
sns.heatmap(pv,annot=True,fmt='d',cmap='GnBu')
# fmt : format -- d : integer
# annot :True -- it shows value at cell present or not
```

# Working with Pandas DataFrames

```python
import pandas as pd

df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.scatterplot(x="age", y="salary", hue="age", data=df)
plt.title('Salary vs Age Scatter Plot')
plt.grid()
plt.show()
```
