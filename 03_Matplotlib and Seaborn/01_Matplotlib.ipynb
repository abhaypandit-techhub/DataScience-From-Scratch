# without using plt.legend()

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
Rohitsharma = [250, 670, 800, 1550, 1500, 1300, 1000, 1750, 950, 1200]

plt.plot(years, kohli, label="Virat Kohli")#label used for legend char in the information
plt.plot(years, sehwag, label="Virender Sehwag")
plt.plot(years, Rohitsharma, label="Rohit sharma")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.show()
```

# Best practice

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
Rohitsharma = [250, 670, 800, 1550, 1500, 1300, 1000, 1750, 950, 1200]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")
plt.plot(years, Rohitsharma, label="Rohit sharma")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend() # it shows label in the box
plt.tight_layout()
plt.show()
```

# using format strings

```python

plt.plot(years, kohli,'ro--', label="Virat Kohli")# 'o' circle marker , r is red , -- double dashed line
plt.plot(years, sehwag,'g*:', label="Virender Sehwag") 
# '^' triangle up marker , g for green ,^ triangular marker : dotted line structure
# '*' star marker , g for green , : dotted line structure

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
```

# other way

```python
plt.plot(years,kohli,color="red",linestyle="--",marker='o',label="Virat kohil")
plt.plot(years,sehwag,color="green",linewidth=2,label="sehwag")
plt.legend()
```

# grid

```python
plt.plot(years,kohli,color="red",linestyle="--",linewidth=2,label="Virat kohil")
plt.plot(years,sehwag,color="green",label="sehwag")
plt.grid(True) # it shows chart in jaali like or .grid() this is also true
plt.legend()
```
