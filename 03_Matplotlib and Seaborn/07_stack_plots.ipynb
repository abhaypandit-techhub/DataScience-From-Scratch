```python
import matplotlib.pyplot as plt
```

```python
activities=["studying","playing","waching tv","sleeping"]
time_spent=[3,2,2,5]
colors=['blue','red','green','pink']
plt.figure(figsize=(6,6))
plt.pie(time_spent,labels=activities,colors=colors,startangle=90,autopct='%1.1f%%')
plt.title("Student activities")
plt.axis('equal') # pie is perfect circle 
```

# stack plot
--legend() parameter loc='best' matplotlib decide itself and loc='upper left'

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
labels=['studyig','playing','watching_tv','sleeping']
colors=['blue','red','green','yellow']

plt.figure(figsize=(6,6))
plt.stackplot(days,studying,playing,watching_tv,sleeping,labels=labels,colors=colors,alpha=0.59,edgecolor='black')
# alpha = 1 → fully opaque (poori tarah visible)
# alpha = 0 → fully transparent (invisible)
# 0 <= alpha <= 1
plt.legend(loc='best') # parameter tells location of the legend,best means matplotlib decide
# plt.legend(loc='upper center')
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Time spent by student in week")
plt.grid() # sometime grid used to see the value
```
