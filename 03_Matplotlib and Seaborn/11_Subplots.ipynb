```python
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
# Data 
```

```python
# plot number -1
plt.subplot(1,2,1) # 1 row 2 column and plot at 1 position 
plt.plot(x,y1)
plt.title("Double of number x")
plt.xlabel("Number")
plt.ylabel("Double of number")
plt.grid()

# plot number -2
plt.subplot(1,2,2)
plt.title("Square of number x")
plt.xlabel("Number")
plt.ylabel("Square of number")
plt.plot(x,y2)
plt.tight_layout()
plt.grid()
```

# # 2x2 grid plot 
This lays out 4 plots in a 2x2 grid. plt.tight_layout() avoids overlapping titles and labels.

```python
# More variations of x
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]

plt.figure(figsize=(8, 6))  # Optional: make it bigger

# 2x2 grid plot form
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('x * 2')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('x squared')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('sqrt(x)')

labels=['A','B','C']
runs=[56,78,90]
explode=[0,0,0.18]
plt.subplot(2, 2, 4)
plt.pie(runs,labels=labels,autopct='%1.1f%%',shadow=True,explode=explode)
# plt.title('10 - x')

plt.tight_layout()
plt.show()
```

```python
fig, axs=plt.subplots(1,2,figsize=(10,4)) # 1row 2column in that case axs is 1D array
axs[0].plot(x,y1)
axs[0].set_title("x*2")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")

axs[1].plot(x,y2)
axs[1].set_title("x**2")

fig.suptitle("My comparision plot",fontsize=15)
fig.tight_layout()
```

```python
study_hours = [2,3,4,5,6,7,8,9]
gpa = [6.1,6.5,6.8,7.2,7.6,8.0,8.4,8.8]
ai_usage = [1,2,3,4,5,6,7,8]
productivity = [55,60,63,68,72,78,85,90]
```

```python
fig , axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].scatter(study_hours,gpa,c=gpa,cmap='BuGn')
axs[0,0].set_xlabel("Hours")
axs[0,0].set_ylabel("GPA score")
axs[0,0].set_title("Study hours vs gpa score")
for i in range(len(study_hours)):
    axs[0,0].annotate(f"s{i+1}",(study_hours[i]-0.4,gpa[i]))

axs[0,0].grid()

axs[0,1].bar(ai_usage,productivity)
axs[0,1].set_xlabel("AI_usage hours")
axs[0,1].set_ylabel("GPA score")
axs[0,1].set_title("AI_usage vs productivity")
axs[0,1].grid()

fig.suptitle("Comparision Graph")
fig.tight_layout()
```

# . Looping Over Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

for i in range(2):
    for j in range(2):
        idx = i * 2 + j
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()
```
