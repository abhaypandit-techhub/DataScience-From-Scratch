# Assignment:
Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
Then create a horizontal bar chart showing total runs scored in their debut 5 years.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [1, 2, 3, 4, 5] 
players = ["S","K","Seh"]
sachin = [450,780,1100,1350,1600]
kohli  = [520,890,1250,1480,1750]
sehwag = [600,950,1200,1400,1550]

y=np.arange(len(years)) 
height=0.20
plt.barh(y+height,sachin,color='skyblue',height=height,label="Sachin Tendulkar",edgecolor='black')
for i in range(len(years)):
    plt.text(sachin[i]+10, y[i]+height,"S",va='center')
plt.barh(y,kohli,color='green',height=height,label="Virat Kohli",edgecolor='black')
for i in range(len(years)):
    plt.text(kohli[i]+5, y[i],"Koh",va='center')
plt.barh(y-height,sehwag,color='red',height=height,label="Virender Sehwag",edgecolor='black')
for i in range(len(years)):
    plt.text(sehwag[i]+5, y[i]-height,"Seh",va='center')

plt.xlabel("Runs score")
plt.ylabel("Years")
plt.title("Comparing the players")
plt.yticks(y,years)
plt.legend(loc='lower right')
plt.grid()
plt.tight_layout()
```
