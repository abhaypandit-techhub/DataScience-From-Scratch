# Assignment:
Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.


```python
import matplotlib.pyplot as plt
```

```python
labels = ["Android", "iOS", "Others"]
market_share = [72, 26, 2]

explode=[0.20,0,0]
plt.pie(market_share,labels=labels,explode=explode,shadow=True,autopct='%1.1f%%',startangle=60)
plt.title("Market share of mobile OS")
```

# comparing vertical bar chart

```python
plt.barh(labels,market_share,edgecolor='black')
plt.xlabel("Market share ")
plt.ylabel("Mobile OS")
plt.title("Market share of mobile OS")
plt.tight_layout()
plt.grid()
```

 pie chart might easier to understand
